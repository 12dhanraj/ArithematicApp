## Load and prepare data

### Subtask:
Define the data path and create a YAML configuration file for the dataset.

**Reasoning**:
Define the path to the unzipped dataset and create a `data.yaml` file that contains the necessary information for YOLOv8 to load and use the dataset, such as the paths to the training and validation images and labels, the number of classes, and the class names.

```python
import os

# Define the path to the unzipped dataset
dataset_path = '/content/rust_dataset'

# Create a data.yaml file for YOLOv8
yaml_content = f"""
train: {dataset_path}/train/images
val: {dataset_path}/valid/images
test: {dataset_path}/test/images

nc: 1  # Number of classes
names: ['rust']  # Class names
"""

# Write the content to a data.yaml file
with open('data.yaml', 'w') as f:
    f.write(yaml_content)

print("data.yaml file created successfully.")
```

## Define and configure the model

### Subtask:
Define the YOLOv8 model and configure training parameters.

**Reasoning**:
Import the `YOLO` class from the `ultralytics` library and define a YOLOv8s model. Then, configure the training parameters such as the number of epochs, image size, and the path to the data configuration file.

```python
from ultralytics import YOLO

# Load a pretrained YOLOv8s model
model = YOLO('yolov8s.pt')

# Configure training parameters
# You can adjust these parameters based on your needs
epochs = 50  # Number of training epochs
imgsz = 640  # Image size
data_config = 'data.yaml' # Path to the data configuration file

print(f"YOLOv8s model loaded and configured for training with {epochs} epochs and image size {imgsz}.")
```

## Train the model

### Subtask:
Train the configured YOLOv8 model on the dataset.

**Reasoning**:
Use the `train()` method of the YOLO model object, providing the path to the data configuration file, the desired number of training epochs, and the image size.

```python
# Train the model
model.train(data=data_config, epochs=epochs, imgsz=imgsz)
```

## Input Phase and Running the Model

### Subtask:
Load the trained model and perform inference on a new image.

**Reasoning**:
Load the best-trained model from the `runs` directory and use its `predict()` method to perform object detection on an example image. This will show how to use the model in a real-world scenario.

```python
from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow
import os
from PIL import Image
import numpy as np

# Load the trained model
# Replace 'runs/detect/train/weights/best.pt' with the actual path to your best model if it's different
model_path = 'runs/detect/train/weights/best.pt'
model = YOLO(model_path)

# --- Input Phase ---
# Provide the path to the image you want to analyze.
# You can upload an image or use one from your Google Drive.
# Example using an image from your Google Drive:
# input_image_path = '/content/drive/My Drive/path/to/your/image.jpg'

# For demonstration, let's assume you have an image named 'test_image.jpg' in your current directory
# You might need to upload a test image or adjust the path accordingly.
input_image_path = '/content/drive/MyDrive/ise/rust.jpg' # Replace with your image path

# Check if the input image exists
if not os.path.exists(input_image_path):
    print(f"Error: Input image not found at {input_image_path}")
else:
    # --- Run Step-by-Step ---

    # Step 1: Load the image
    print(f"Step 1: Loading image from {input_image_path}")
    img = cv2.imread(input_image_path)

    if img is None:
        print(f"Error: Could not load image from {input_image_path}")
    else:
        # Step 2: Perform inference
        print("Step 2: Performing inference with the trained model")
        results = model.predict(img)

        # Step 3: Process and display results
        print("Step 3: Processing and displaying results")
        # The results object contains the detections
        # You can access bounding boxes, confidence scores, and class labels

        # Assuming you want to display the image with detections
        for r in results:
            im_array = r.plot()  # plot a BGR numpy array of predictions
            im = Image.fromarray(im_array[..., ::-1])  # RGB PIL image
            # Display the image with detections
            cv2_imshow(cv2.cvtColor(np.array(im), cv2.COLOR_RGB2BGR))

        print("Inference complete. Detections displayed above.")
```
